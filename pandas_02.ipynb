# Dropna Function and It's Parameter

````python
import pandas as pd
import numpy as np

# Creating a sample DataFrame with missing values
data = {
    'A': [1, 2, np.nan, 4],
    'B': [np.nan, 2, np.nan, 4],
    'C': [1, np.nan, np.nan, 4],
    'D': [np.nan, np.nan, np.nan, np.nan]
}
df = pd.DataFrame(data)

# Displaying the original DataFrame
print("Original DataFrame:")
print(df)

# Example 1: Drop rows with any missing values
df_drop_any = df.dropna(axis=0, how='any')
print("\nDataFrame after dropping rows with any missing values:")
print(df_drop_any)

# Example 2: Drop columns with all missing values
df_drop_all = df.dropna(axis=1, how='all')
print("\nDataFrame after dropping columns with all missing values:")
print(df_drop_all)

# Drop rows with less than 2 non-NA values
df_thresh = df.dropna(axis=0, thresh=3)
print("\nDataFrame after dropping rows with less than 3 non-NA values:")
print(df_thresh)

# Drop rows, but only consider subset of columns 'A' and 'C'
df_subset = df.dropna(axis=0, subset=['A', 'C'])
print("\nDataFrame after dropping rows considering subset of columns 'A' and 'C':")
print(df_subset)
````

# Fillina Funtion And It's Parameter

````python
import pandas as pd
import numpy as np

# create a sample DataFrame with missing values
data = {'A': [1, np.nan, np.nan, 4], 
        'B': [5, np.nan, 7, 8], 
        'C': [9, np.nan, 11, np.nan]}
df = pd.DataFrame(data)

print("Original DataFrame:")
print(df)

# fill missing values with a scalar value
df_filled = df.fillna(2450)
print("\nFilled DataFrame with scalar value 2450:")
print(df_filled)

# fill missing values with a dictionary
fill_values = {'A': 200, 'B': 300, 'C': 400}
df_filled = df.fillna(fill_values)
print("\nFilled DataFrame with dictionary A=> 200, B=> 300, C=> 400")
print(df_filled)

# Fill NaNs using forward fill
df_filled = df.ffill(axis=0)  # fill along rows
print("\nFilled DataFrame along rows (forward fill) :")
print(df_filled)

# Fill NaNs using backward fill
df_filled = df.bfill(axis=1)  # fill along columns
print("\nFilled DataFrame along columns (backward fill) :")
print(df_filled)

# fill missing values with a limit 1 along with rows
df_filled = df.fillna(2450, limit=1, axis=0)  # fill only the first consecutive NaN value
print("\nFilled DataFrame with limit 1 along with rows:")
print(df_filled)

# fill missing values in place
df.fillna(2450, inplace=True)
print("\nChanges occurs in original DataFrame:")
print(df)
````

# Replace Funtion

````python
# Sample DataFrame 
data = {
    'A': [1, 2, 3, 4, 5],
    'B': [10, 20, 30, 40, 50],
    'C': ['a', 'b', 'c', 'd', 'e']
}
df = pd.DataFrame(data)
print(f'Original DataFrame \n {df}')

print('-----------------------------')

# Replace specific value
# replaced_df_value= df.replace(to_replace=10, value=100)
# OR   Can Use Following Code
replaced_df_value = df.replace(10, 100)
print(f'Replaced Specific Value 10 Replaced By 100 \n {replaced_df_value}')

print('-----------------------------')

# Replace multiple values
replaced_df_multiple_values = df.replace([1, 20], [100, 200])
print(f'Replaced Multiple Values 1 And 20 Replaced By 100 And 200 \n {replaced_df_multiple_values}')

print('-----------------------------')

# Replace using a dictionary
replaced_df_dict = df.replace({'A': 1, 'B': 10}, {'A': 100, 'B': 1000})
print(f'Replaced Value from Specific Column A: 1 And B: 10 Replace By A: 100 And B: 1000 \n {replaced_df_dict}')

print('-----------------------------')

# Replace using nested dictionaries for specific column
replaced_df_nested_dict = df.replace({'A': {1: 100, 2: 200}, 'B': {10: 1000, 20: 2000}})
print(f"Replaced DataFrame (nested dictionary) A: 1,2 And B: 10,20 Replaced By A: 100,200 And B: 1000,2000")
print(replaced_df_nested_dict)
````

````python
data = pd.DataFrame({
    'A': ['foo', 'fool', 'foobar', 'bar', 'baz'],
    'B': ['foo', 'bar', 'baz', 'qux', 'quux']
})

print(f'Original DataFrame \n {data}')

print('-----------------------------')

# This will replace all occurrences of 'foo' with 'foobaz' in the DataFrame.
replaced_df_regex = data.replace(to_replace='foo', value='foobaz', regex=True)
print(f"Replaced DataFrame (regex) foo Replaced By foobaz \n {replaced_df_regex}")

print('-----------------------------')

# You Can See Original df Is Not Changing Now I Will Change Original df With inplace Parameter
data.replace(to_replace='foo', value='python', inplace=True)
print(f"Changes occurs in original DataFrame \n {data}")
````

# Interpolation Function
- interpolation is a method used to fill in missing data in a DataFrame or Series. And It Just Fill Numeric dtype

#### **You Can Pass Following Values In Method Parameter By Default linear**
```python
method : { linear, time, index, values, nearest, zero, slinear, quadratic, cubic, barycentric, 
krogh, polynomial, spline, piecewise_polynomial, from_derivatives, pchip, akima }
```

````python
import pandas as pd
import numpy as np

# Sample data with missing values
data = {
    'A': [1, np.nan, np.nan, 2, np.nan, 6, np.nan],
    'B': [np.nan, 2, np.nan, 4, 5, np.nan, 7]
}
df = pd.DataFrame(data)

# Print the original DataFrame
print("Original DataFrame:")
print(df)

# Interpolate missing values
df_interpolated = df.interpolate(method='linear', limit_area=None, axis=0, limit=2, limit_direction='both')

# Print the interpolated DataFrame
print("\nInterpolated DataFrame:")
print(df_interpolated)
````

# Merge Function

````python
import pandas as pd

# Create sample DataFrames
df1 = pd.DataFrame({
    'key': ['A', 'B', 'C', 'D'],
    'value': [1, 2, 3, 4]
})

df2 = pd.DataFrame({
    'key': ['A', 'B', 'E', 'F'],
    'value': [5, 6, 7, 8]
})

# Merging DataFrames
merged_df = pd.merge(
    df1, df2, how='left', left_index=True, right_index=True, indicator=False , suffixes=('_1', '_2') )

# Display the merged DataFrame
merged_df
````

# Concat Function

````python
import pandas as pd

# Create sample DataFrames
df1 = pd.DataFrame({
    'A': ['A0', 'A1', 'A2', 'A3'],
    'B': ['B0', 'B1', 'B2', 'B3']
})

df2 = pd.DataFrame({
    'A': ['A4', 'A5', 'A6'],
    'B': ['B4', 'B5', 'B6']
})

# Concatenating DataFrames
concatenated_df = pd.concat(
    [df1, df2], axis=1, join='outer', keys=['Data_1', 'Data_2'], ignore_index=False)

# Display the concatenated DataFrame
concatenated_df
````

# Join Function

````python
import pandas as pd

# Create two sample DataFrames
df1 = pd.DataFrame({
    'A': [1, 2, 3],
    'B': [4, 5, 6]
}, index=['a', 'b', 'c'])

df2 = pd.DataFrame({
    'A': [1, 2, 7],
    'D': [10, 11, 12]
}, index=['a', 'b', 'f'])

# Perform a join Method
# If Column Name Is Different Of Both DF So There Is No Need To Pass Parameter of lsuffix Or rsuffix
result = df1.join(df2, how='outer', lsuffix='_x', rsuffix='_y')

# show result
result
````

# Groupby Function

````python
import pandas as pd

# Create a DataFrame
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve', 'Frank'],
    'Department': ['HR', 'IT', 'IT', 'HR', 'HR', 'IT'],
    'Salary': [70000, 80000, 75000, 72000, 69000, 85000]
}

# Convert dictionary to DataFrame
df = pd.DataFrame(data)

# Display the DataFrame
print("Original DataFrame: ")
print(df)

print('\n------------------------------------------\n')

# Group the DataFrame by 'Department'
# The groupby method returns a GroupBy object, which contains the split data
grouped= df.groupby('Department')

# Display the groupby object detail
print(f'Groupby Object \n {grouped}')

print('\n------------------------------------------\n')

# Iterate Over Each Group
print('Group the DataFrame by Department')
for department ,grouped_data in grouped:
    print(department)
    print(grouped_data)

print('\n------------------------------------------\n')

# Getting data for the group 'HR'  
HR_group = grouped.get_group('HR')  
print(f'Getting data for the group HR \n {HR_group}')
````

# Melt Function

### Melt Method: 
- The melt() method in pandas is used to transform a DataFrame from a wide format into a long format

````python
import pandas as pd

# sample data
data={'Days': [1, 2, 3], 'Name': ['Ammar', 'Karan', 'Amir'], 'Eng': [10, 20, 30], 'Math': [5, 10, 15],
      'Science': [15, 20, 30], 'History': [30, 50, 60] }

# create a DataFrame
df= pd.DataFrame(data)

print(f'Original DataFrame \n {df}')

print('\n---------------------------------\n')

# Let's Transform The DF
print('Transformed DF')

pd.melt(df, id_vars=['Days', 'Name'], value_vars=['Math', 'Eng', 'Science'], value_name='Marks', 
        var_name='Subjects')
````
